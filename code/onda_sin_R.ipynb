```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import fig_size

%matplotlib widget

width = 469.75502
graph_path = '../assets/'
data_path = '../data/'
table_path = '../tables/'

mpl.rcParams.update({
    # Use LaTeX to write all text
    "text.usetex": True,
    "font.family": "serif",
    # Use 10pt font in plots, to match 10pt font in document
    "axes.labelsize": 11,
    "font.size": 11,
    # Make the legend/label fonts a little smaller
    "legend.fontsize": 9,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
})
```

```python
df: pd.DataFrame = pd.read_csv(data_path + 'V_R.csv', sep=r'\s*,\s*', engine='python')
df = df.sort_values('f')
df
```

### Errori su $V_{in}$ e $V_{out}$

```python
df['dV_in'] = df['V_in/sc'] / 10
df['dV_out'] = df['V_out/sc'] / 10
df
```

### Errori su $t$

```python
df['dt'] = df['t/sc'] / 10
df
```

```python
df['|A|'] = df['V_out'] / df['V_in']
eV_in = df['dV_in'] / df['V_in']
eV_out = df['dV_out'] / df['V_out']
df['d|A|'] = df['|A|'] * (eV_in + eV_out)
df
```

```python
columns = ['V_in', 'dV_in', 'V_out', 'dV_out', 'f', 't', 'dt']

if False:
    df.to_latex(
        buf = table_path + 'onda_sin_VR.tex',
        columns = columns,
        header = [r'$V_{in} \; (V)$', r'$\delta V_{in} \; (V)$', r'$V_{out} \; (V)$', r'$\delta V_{out} \; (V)$', r'$f \; (kHz)$', r'$t \; (\mu s)$', r'$\delta t \; (\mu s)$'],
        index = False,
        bold_rows = True,
        column_format = 'cc|cc|ccc',
        escape = False,
    )
d = df.loc[:, columns]
d
```

```python
columns = ['|A|', 'd|A|', 'f']

if False:
    df.to_latex(
        buf = table_path + 'onda_sin_AR.tex',
        columns = columns,
        header = [r'$|A_{R}|$', r'$d|A_{R}|$', r'$f \; (kHz)$'],
        index = False,
        bold_rows = True,
        column_format = 'cc|c',
        escape = False,
    )
d = df.loc[:, columns]
d
```

```python
fig, ax = plt.subplots(figsize = fig_size.from_pt(width))
# fig.tight_layout()

ax.set_xscale('log')
ax.set_yscale('log')

ax.errorbar(
    x = df['f'],
    y = df['|A|'], yerr=df['d|A|'],
    fmt='o', markersize = 3,
    capsize = 3, label='Misure con errori'
)

xlim = np.array(ax.get_xlim())

# |A| = 1/sqrt(2)
# A_taglio = lambda x: x - x +(1/2)**(1/2)
# ax.plot(xlim, A_taglio(xlim), linestyle='dashed', lw=0.8, label=r'$1/\sqrt{2}$')

ax.set_xlim(xlim)

ax.minorticks_on()
ax.grid(True, which='major', lw = 1)
ax.grid(True, which='minor', lw = 0.5)

ax.legend()

ax.set_xlabel(r'$f \; (kHz)$')
ax.set_ylabel(r'$|A_{R}|$')

fig.savefig(graph_path + 'onda_sin_AR(f).pdf', format='pdf')
```

```python
df['phi'] = - 2 * np.pi * df['t'] / df['T'] + np.pi/2
et = df['dt'] / df['t']
df['dphi'] = df['phi'] * et
df
```

```python
columns = ['phi', 'dphi', 'f']

if False:
    df.to_latex(
        buf = table_path + 'onda_sin_phiR.tex',
        columns = columns,
        header = [r'$\varphi_{R} \; (rad)$', r'$d\varphi_{R} \; (rad)$', r'$f \; (kHz)$'],
        index = False,
        bold_rows = True,
        column_format = 'cc|c',
        escape = False,
    )
d = df.loc[:, columns]
d
```

```python
fig, ax = plt.subplots(figsize = fig_size.from_pt(width))
# fig.tight_layout()

ax.set_xscale('log')

ax.errorbar(
    x = df['f'],
    y = df['phi'], yerr=df['dphi'],
    fmt='o', markersize = 3,
    capsize = 3, label='Misure con errori'
)

xlim = np.array(ax.get_xlim())
ylim = np.array(ax.get_ylim())

phi_taglio = lambda x: x - x + np.pi/4
ax.plot(xlim, phi_taglio(xlim), lw=0.8, label=r'$\pi/4$')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

ax.minorticks_on()
ax.grid(True, which='major', lw = 1)
ax.grid(True, which='minor', lw = 0.5)

ax.legend()

ax.set_xlabel(r'$f \; (kHz)$')
ax.set_ylabel(r'$\varphi_{R} \; (rad)$')

fig.savefig(graph_path + 'onda_sin_phi(f)_R.pdf', format='pdf')
```
