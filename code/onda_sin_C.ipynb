```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import fig_size
import lab_round

%matplotlib widget

width = 469.75502
graph_path = '../assets/'
data_path = '../data/'
table_path = '../tables/'

mpl.rcParams.update({
    # Use LaTeX to write all text
    "text.usetex": True,
    "font.family": "serif",
    # Use 10pt font in plots, to match 10pt font in document
    "axes.labelsize": 11,
    "font.size": 11,
    # Make the legend/label fonts a little smaller
    "legend.fontsize": 9,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
})
```

```python
df: pd.DataFrame = pd.read_csv(data_path + 'V_C.csv', sep=r'\s*,\s*', engine='python')
df = df.sort_values('f')
df
```

### Errori su $V_{in}$ e $V_{out}$

```python
df['dV_in'] = df['V_in/sc'] / 10
df['dV_out'] = df['V_out/sc'] / 10
df
```

### Errori su $t$

```python
df['dt'] = df['t/sc'] / 10
df
```

### Calcolo del modulo della funzione di trasferimento

```python
df['|A|'] = df['V_out'] / df['V_in']
eV_in = df['dV_in'] / df['V_in']
eV_out = df['dV_out'] / df['V_out']
df['d|A|'] = df['|A|'] * (eV_in + eV_out)
df
```

```python
df_str = df.apply(
    lambda x: pd.Series(
        lab_round.round_measure(x['V_in'], x['dV_in']),
        index=['V_in', 'dV_in']
    ), axis=1
).join(
    df.apply(
        lambda x: pd.Series(
            lab_round.round_measure(x['V_out'], x['dV_out']),
            index=['V_out', 'dV_out']
        ), axis=1
    )
).join(
    df.apply(
        lambda x: pd.Series(
            lab_round.round_sig(x['f'], digits=4, one_more_digit=''),
            index=['f']
        ), axis=1
    )
).join(
    df.apply(
        lambda x: pd.Series(
            lab_round.round_measure(x['t'], x['dt']),
            index=['t', 'dt']
        ), axis=1
    )
)

if False:
    df_str.to_latex(
        buf = table_path + 'onda_sin_VC.tex',
        header = [r'$V_{in} \; (V)$', r'$\delta V_{in} \; (V)$', r'$V_{out} \; (V)$', r'$\delta V_{out} \; (V)$', r'$f \; (kHz)$', r'$t \; (\mu s)$', r'$\delta t \; (\mu s)$'],
        index = False,
        bold_rows = True,
        column_format = 'cc|cc|ccc',
        escape = False,
    )

df_str
```

```python
df_str = df.apply(
    lambda x: pd.Series(
        lab_round.round_measure(x['|A|'], x['d|A|']),
        index=['|A|', 'd|A|']
    ), axis=1
).join(
    df.apply(
        lambda x: pd.Series(
            lab_round.round_sig(x['f'], digits=4, one_more_digit=''),
            index=['f']
        ), axis=1
    )
)

if False:
    df_str.to_latex(
        buf = table_path + 'onda_sin_AC.tex',
        header = [r'$|A_{C}|$', r'$d|A_{C}|$', r'$f \; (kHz)$'],
        index = False,
        bold_rows = True,
        column_format = 'cc|c',
        escape = False,
    )

df_str
```

```python
fig, ax = plt.subplots(figsize = fig_size.from_pt(width))

ax.set_xscale('log')
ax.set_yscale('log')

ax.errorbar(
    x = df['f'],
    y = df['|A|'], yerr=df['d|A|'],
    fmt='o', markersize = 3,
    capsize = 3, label='Misure con errori'
)

xlim = np.array(ax.get_xlim())

# |A| = 1/sqrt(2)
# A_taglio = lambda x: x - x +(1/2)**(1/2)
# ax.plot(xlim, A_taglio(xlim), linestyle='dashed', lw=0.8, label=r'$1/\sqrt{2}$')

ax.set_xlim(xlim)

ax.minorticks_on()
ax.grid(True, which='major', lw = 1)
ax.grid(True, which='minor', lw = 0.5)

ax.legend()

ax.set_xlabel(r'$f \; (kHz)$')
ax.set_ylabel(r'$|A_{C}|$')

fig.savefig(graph_path + 'onda_sin_AC(f).pdf', format='pdf')
```

```python
df['phi'] = - 2 * np.pi * df['t'] * df['f'] / 1000
# et = df['dt'] / df['t']
# df['dphi'] = df['phi'] * et
df['dphi'] = 2 * np.pi * df['f'] * df['dt'] / 1000
df
```

```python
df_str = df.apply(
    lambda x: pd.Series(
        lab_round.round_measure(-x['phi'], x['dphi']),
        index=['phi', 'dphi']
    ), axis=1
).join(
    df.apply(
        lambda x: pd.Series(
            lab_round.round_sig(x['f'], digits=4, one_more_digit=''),
            index=['f']
        ), axis=1
    )
)

if False:
    df_str.to_latex(
        buf = table_path + 'onda_sin_phiC.tex',
        header = [r'$\varphi_{C} \; (rad)$', r'$d\varphi_{C} \; (rad)$', r'$f \; (kHz)$'],
        index = False,
        bold_rows = True,
        column_format = 'cc|c',
        escape = False,
    )

df_str
```
